```python
import pandas as pd
import numpy as np
import datetime as dt
df_agent_logging_report = pd.read_csv(r"C:\Users\sirig\Downloads\AgentLogingReport.csv")
df_agent_performance=pd.read_csv(r"C:\Users\sirig\Downloads\AgentPerformance.csv")
```

```python
#1 .Find out their average rating on weekly basis keep this in a mind that they take two days of leave
df_agent_performance.dtypes
df_agent_performance['Date']=pd.to_datetime(df_agent_performance['Date'])
df_agent_performance['week']=df_agent_performance['Date'].apply(lambda x: x.isocalendar().week)
df_non_holiday=df_agent_performance[df_agent_performance['Average Rating']!=0]
agents_avg_rating_weekly_basis=pd.DataFrame(df_non_holiday.groupby(['Agent Name','week']).agg(average_weekly_rating=('Average Rating','mean')))
agents_avg_rating_weekly_basis.head(10)
```

```python
#2 .Total working days for each agents
df_agent_logging_report['Date']=df_agent_logging_report['Date'].astype('datetime64[ns]')
df_agent_logging_report['day']=df_agent_logging_report['Date'].dt.day
df_agent_logging_report.groupby('Agent').agg(working_days=('day','count')).head(10)
```

```python
#3. Total query that you have taken
agent_total_queries_done=df_agent_performance.groupby('Agent Name').agg(total_queries_done=('Total Chats','sum'))
agent_total_queries_done.head(10)
```

```python
#4. total Feedback that you have received
agent_total_feedback_received =df_agent_performance.groupby('Agent Name').agg(total_feedback_received=('Total Feedback','sum'))
agent_total_feedback_received.head(10)
```

```python
#5. a agent name who have average rating between 3.5 to 4
agents_avg_rating_weekly_basis [(agents_avg_rating_weekly_basis['average_weekly_rating']>3.5) & (agents_avg_rating_weekly_basis['average_weekly_rating'] <4)]
```

```python
#6 . Agent name who have rating less then 3.5
agents_avg_rating_weekly_basis [(agents_avg_rating_weekly_basis['average_weekly_rating']<3.5)]
```

```python
#7 . agent name who have rating more then 4.5 average
rat_more_than_45=agents_avg_rating_weekly_basis [(agents_avg_rating_weekly_basis['average_weekly_rating']>4.5)]
rat_more_than_45.head(5)
```

```python
#8 . how many feedback, agents have received more then 4.5 average
feedback_more_than_45_weekly_count = rat_more_than_45.groupby('Agent Name').count()
feedback_more_than_45_weekly_count.head(10)
```

```python
#9 . average weekly response time for each agent
df_agent_performance['Average Response Time']=df_agent_performance['Average Response Time'].astype('timedelta64[s]')
df_agent_performance['Average Response Time in seconds']=df_agent_performance['Average Response Time'].dt.total_seconds()
avg_weekly_response=pd.DataFrame(df_agent_performance.groupby('Agent Name')['Average Response Time in seconds'].mean())
avg_weekly_response.head(15)
```

```python
#10 . average weekly resolution time for each agents
df_agent_performance['Average Resolution Time']=df_agent_performance['Average Resolution Time'].astype('timedelta64[s]')
df_agent_performance['Average Resolution Time in seconds']=df_agent_performance['Average Resolution Time'].dt.total_seconds()
avg_weekly_resolution=pd.DataFrame(df_agent_performance.groupby('Agent Name')['Average Resolution Time in seconds'].mean())
avg_weekly_resolution.head(10)
```

```python
#11 . list of all agents name
len(df_agent_performance['Agent Name'].unique())
pd.DataFrame(df_agent_performance['Agent Name'].unique(),columns=['agent_name_unique'])
```

```python
#12 . percentage of chat on which they have received a feedback
dfs=df_agent_performance.groupby('Agent Name')[['Total Chats','Total Feedback']].sum()
dfs=dfs[(dfs['Total Chats']>0)&(dfs['Total Feedback']>0)] #delete all with zero chat and zero feedback
dfs['percentage feedback']= (dfs['Total Feedback']/dfs['Total Chats'])*100
dfs.head(10)
```

```python
#13 . Total contribution hour for each and every agents weekly basis
df_agent_logging_report['Duration']=df_agent_logging_report['Duration'].astype('timedelta64[s]')
df_agent_logging_report['Duration_in_seconds']=df_agent_logging_report['Duration'].dt.total_seconds()
df_agent_logging_report['Duration_in_hours']=df_agent_logging_report['Duration_in_seconds']/3600
pd.DataFrame(df_agent_logging_report.groupby(['Agent','week'])['Duration_in_hours'].sum()).head(10)
```

```python
#14. total percentage of active hour for a month
dfo=pd.DataFrame(df_agent_logging_report.groupby(['Agent'])['Duration_in_hours'].sum()).head(10)
dfo['total active percentage per month']= (dfo.Duration_in_hours/192)*100
dfo
```
