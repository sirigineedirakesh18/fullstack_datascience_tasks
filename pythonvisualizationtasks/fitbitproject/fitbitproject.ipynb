```python
import pandas as pd
import numpy as np
import datetime as dt
#importing all the modules that are required
```

```python
#1.reading the dataset 
df_fit_bit=pd.read_csv(r"C:\Users\sirig\Downloads\FitBit data.csv")
df_fit_bit.head(5)
```

```python
#Convert all date columns in the dataset to timestamp format in Pandas
df_fit_bit['ActivityDate']=df_fit_bit['ActivityDate'].astype('datetime64[ns]')
```

```python
#Find the total number of unique IDs in the dataset.
df_fit_bit['Id'].nunique()
```

```python
#Identify the most active ID in the dataset.
df_fit_bit.groupby('Id')['Calories'].sum()[df_fit_bit.groupby(['Id'])['Calories'].sum()==max(df_fit_bit.groupby('Id')['Calories'].sum())].index
```

```python
#Count how many IDs have never logged any activity.
zero_act_id=pd.DataFrame(df_fit_bit[df_fit_bit['LoggedActivitiesDistance']==0]['Id'].unique())
activity_logged=pd.DataFrame(df_fit_bit[df_fit_bit['LoggedActivitiesDistance']!=0]['Id'].unique())
pd.DataFrame(set(zero_act_id[0])-set(activity_logged[0]))
#these are the ids that never logged activity
```

```python
#Identify the laziest person (ID) in the dataset.
df_fit_bit.groupby('Id')['Calories'].sum()[df_fit_bit.groupby(['Id'])['Calories'].sum()==min(df_fit_bit.groupby('Id')['Calories'].sum())].index

     
```

```python
#Research the recommended calorie burn for a healthy person and determine how many individuals in the dataset meet this criterion.
df_fit_bit[df_fit_bit['Calories']>2200].head(5)
```

```python
#Identify individuals who are not regular in their activities.
ser=df_fit_bit.groupby(['Id'])['Id'].count()<15
ser[ser==True].index.to_list()

```

```python
# Define the threshold for regular activity
threshold = 15

# Group by 'Id' and count the number of days each individual has logged activities
activity_counts = df_fit_bit.groupby('Id')['ActivityDate'].count()

# Identify individuals who are not regular in their activities
not_regular_ids = activity_counts[activity_counts <= threshold].index
not_regular_ids
not_regular_df = pd.DataFrame({'Id': not_regular_ids, 'Count': activity_counts[not_regular_ids].values})
not_regular_df
```

```python
#Find the third most active person in the dataset using both Pandas
total_cal=df_fit_bit.groupby('Id')['Calories'].sum()
total_cal.sort_values(ascending=False).index[2]
```

```python
#Identify the fifth laziest person in the dataset
total_cal=df_fit_bit.groupby('Id')['Calories'].sum()
total_cal.sort_values(ascending=True).index[4]
```

```python
#Calculate the total cumulative calories burned for each person
pd.DataFrame(df_fit_bit.groupby(['Id'])['Calories'].sum())
```
